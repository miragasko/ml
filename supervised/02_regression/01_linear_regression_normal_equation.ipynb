```python
import numpy as np
```

```python
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'years of work: {X1}')
print(f'reward: {Y}')
print(f'number of samples: {m}')
```

```python
X1 = X1.reshape(m, 1)
print(X1)
print(X1.shape)
```

```python
bias = np.ones((m,1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

image.png

```python
np.dot(X.T, X)
```

```python
L = np.linalg.inv(np.dot(X.T, X))
L #left side
```

```python
P = np.dot(X.T, Y)
P
```

```python
L * P # just a test - do not consider this product
```

```python
test = L @ P
test
```

```python
np.dot(L, P)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X1, Y)

print(regression.intercept_)
print(regression.coef_[0])
```
